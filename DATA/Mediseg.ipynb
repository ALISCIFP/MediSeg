```python
%matplotlib inline

import matplotlib
import numpy as np
import matplotlib.pyplot as plt
```

```python
import scipy.misc
import skimage.color
for i in range(1, 10):
    fig = plt.figure(figsize=(18,10))
    gt_file = "OP4/img_0%i.png" % i
    ax1 = fig.add_subplot(1, 5, 1)
    ax2 = fig.add_subplot(1, 5, 2)
    ax3 = fig.add_subplot(1, 5, 3)
    ax4 = fig.add_subplot(1, 5, 4)
    ax5 = fig.add_subplot(1, 5, 5)
    orig = scipy.misc.imread(gt_file)
    ax1.imshow(orig)
    ax2.imshow(skimage.color.rgb2hed(orig))
    ax3.imshow(skimage.color.rgb2hsv(orig))
    ax4.imshow(skimage.color.rgb2lab(orig))
    ax5.imshow(skimage.color.rgb2luv(orig))
    plt.show()
```
